**3. Bar chart**

alasan memilih Bar chart dikarenakan kita hanya menghitung 3 jenis makanan yang penjualan nya terendah maka untuk mendapatkan hasil yang sesuai saya akhirnya memilih bar char untuk mengetahui 3 jenis makanan yang terendah

1. dapat dilihat bahwa jenis makanan yang tertinggi adalah ice cream sedangkan yang terendah adalah kue wajik

2. tiga makanan terbawah yang harus disingkirkan adalah **pudding vanila, pastel, dan kue wajik**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Jenis_makanan = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies',
                'Puding Strawberi', 'Puding Coklat', 'Ice Cream Nutela', 'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jumlah_penjualan = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'jenis' :Jenis_makanan, 'Jumlah' : jumlah_penjualan,})
df.sort_values(by='Jumlah', inplace = True, ascending=False)

kor_x = np.arange(len(df))
warna = ['#000080' for _ in range(len(df))]
warna [-3] = '#FF0000'
warna [-2] = '#FF0000'
warna [-1] = '#FF0000'

plt.figure(figsize=(20,10))

plt.bar(kor_x, df['Jumlah'], tick_label=df['jenis'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Jumlah penjualan')
plt.title('Penjualan makanan kafe biru')
plt.show() 
```
